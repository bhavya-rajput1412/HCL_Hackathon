```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
file_path = "use_case1.csv"   # change if your file name is different
df = pd.read_csv(file_path)

df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')
df['Total_Amount'] = pd.to_numeric(df['Total_Amount'], errors='coerce')
```

```python
df.dtypes
```

```python
rules = pd.DataFrame(index=df.index)
rules['not_null_key'] = df['Transaction_ID'].notna() & df['Customer_ID'].notna()
rules['valid_amount'] = df['Amount'].notna() & (df['Amount'] > 0)
rules['valid_date'] = df['Date'].notna()
rules['valid_email'] = df['Email'].str.contains('@', na=False)

valid_mask = rules.all(axis=1)

df_clean = df[valid_mask].copy()
df_bad = df[~valid_mask].copy()
df_bad['error_reason'] = rules.apply(
    lambda r: ','.join(r.index[~r.values]), axis=1
)

print("Clean rows:", len(df_clean))
print("Bad rows:", len(df_bad))
```

```python
engine = create_engine("sqlite:///retail_loyalty.db", echo=False)

# All valid transactions
df_clean.to_sql("transactions", engine, if_exists="replace", index=False)

# Rejected records
df_bad.to_sql("rejected_records", engine, if_exists="replace", index=False)
```

```python
cust_cols = ['Customer_ID','Name','Email','Phone','Address','City','State',
             'Zipcode','Country','Age','Gender','Income','Customer_Segment']

customers = (df_clean[cust_cols].drop_duplicates().rename(columns={'Customer_ID':'customer_id'}))

customers['total_points'] = 0

customers.to_sql("customer_details", engine, if_exists="replace", index=False)
customers.head()
```

```python
df_clean['Total_Amount'] = df_clean['Total_Amount'].fillna(0)

df_clean['points_earned'] = (df_clean['Total_Amount'] // 100).astype(int)
```

```python
df_clean['points_earned'] = (df_clean['Total_Amount'] // 100).astype(int)

loyalty_txn = df_clean[['Transaction_ID','Customer_ID','Date','Total_Amount','points_earned']]

loyalty_txn.to_sql("loyalty_transactions", engine,if_exists="replace", index=False)

loyalty_txn.head()
```

```python
points_by_customer = (loyalty_txn.groupby('Customer_ID')['points_earned'].sum().reset_index().rename(columns={'Customer_ID':'customer_id','points_earned':'total_points'}))

points_by_customer.head()
```

```python
points_by_customer = (loyalty_txn.groupby('Customer_ID')['points_earned'].sum().reset_index().rename(columns={'Customer_ID':'customer_id','points_earned':'total_points'}))
points_by_customer.head()
```

```python
customers = pd.read_sql("SELECT * FROM customer_details", engine)

customers = customers.drop(columns=['total_points'], errors='ignore')
customers = customers.merge(points_by_customer, on='customer_id', how='left')
customers['total_points'] = customers['total_points'].fillna(0).astype(int)

customers.to_sql("customer_details", engine, if_exists="replace", index=False)
customers.head()
```

```python
ref_date = df_clean['Date'].max()

rfm = (df_clean.groupby('Customer_ID').agg({'Date': lambda x: (ref_date - x.max()).days,'Transaction_ID': 'count','Total_Amount': 'sum'}).rename(columns={'Date':'Recency','Transaction_ID':'Frequency','Total_Amount':'Monetary'}).reset_index().rename(columns={'Customer_ID':'customer_id'}))

rfm.to_sql("rfm_metrics", engine, if_exists="replace", index=False)
rfm.head()
```

```python
cust = pd.read_sql("SELECT * FROM customer_details", engine)
rfm = pd.read_sql("SELECT * FROM rfm_metrics", engine)

seg = cust.merge(rfm, on='customer_id', how='left')
seg.head()
```

```python
# High-spender threshold
threshold = seg['Monetary'].quantile(0.9)
seg['is_high_spender'] = seg['Monetary'] >= threshold

# At-risk
seg['is_at_risk'] = (seg['Recency'] > 30) & (seg['total_points'] > 0)

def label_segment(row):
    if row['is_high_spender']:
        return 'High-Spender'
    if row['is_at_risk']:
        return 'At-Risk'
    return 'Regular'

seg['segment'] = seg.apply(label_segment, axis=1)

seg.to_sql("customer_segments", engine, if_exists="replace", index=False)
seg[['customer_id','Monetary','Recency','total_points','segment']].head()
```

```python
from sqlalchemy import text


with engine.begin() as conn:
    conn.exec_driver_sql("DROP TABLE IF EXISTS new_customer")

create_sql = """
CREATE TABLE new_customer (
    prospect_id        TEXT PRIMARY KEY,
    first_txn_date     DATE,
    last_txn_date      DATE,
    txn_count          INTEGER,
    total_spend        REAL,
    avg_order_value    REAL,
    city               TEXT,
    preferred_category TEXT,
    most_used_channel  TEXT
);
"""

with engine.begin() as conn:
    conn.exec_driver_sql(create_sql)
```
